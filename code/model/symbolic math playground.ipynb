```python
import sympy as sp
from sympy import symbols
from sympy import solve
```

```python
t1, t2, ε1, ε2, κ1, κ2, p1, p2, l1, l2, α, w1, w2 = symbols(["t1", "t2", "ε1", "ε2", "κ1", "κ2", "p1", "p2", "l1", "l2", "α", "w1", "w2"])
q1_1, q1_2, q2_1, q2_2, = symbols(["q1_1", "q1_2", "q2_1", "q2_2"])
λ1, λ2, μ1, μ2 = symbols(["λ1", "λ2", "μ1", "μ2"])
λ, μ = symbols(["λ", "μ"])
```

```python
# Solve constraints
solve([(1-t1-t2)*(1-ε1)*κ1*p1**(-ε1)-λ*ε1*κ1*p1**(-1-ε1),
       (1-t1-t2)*(1-ε2)*κ2*p2**(-ε2)-λ*ε2*κ2*p2**(-1-ε2),
       (1-t1-t2)*(-w1)-λ*α*l1**(α-1),
       (1-t1-t1)*(-w2)-λ*α*l2**(α-1),
       κ1*p1**(-ε1)+κ2*p2**(-ε2)-l1**α-l2**α],
     [p1, p2, l1, l2, λ])
```

```python
# Separate accounting optimization conditions
solve([(1-t1)*q1_1+(1-t2)*q2_1-κ1*λ1*ε1*p1**(-ε1-1),
       (1-t1)*q1_2+(1-t2)*q2_2-κ2*λ2*ε2*p2**(-ε2-1),
       -(1-t1)*w1+μ1*α*l1**(α-1),
       -(1-t2)*w2+μ2*α*l2**(α-1),
       (1-t1)*p1-μ1-λ1,
       (1-t2)*p1-μ2-λ1,
       (1-t1)*p2-μ1-λ2,
       (1-t2)*p2-μ2-λ2], [p1,p2,l1,l2,μ1,μ2,λ1,λ2])
```
